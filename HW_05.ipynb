# Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

## Functions (google API)

```python
def getstock(q , startdate , enddate):
    import requests

    url = 'https://finance.google.com/finance/historical'
    
    parameters = {}

    parameters['q'] = q
    parameters['startdate'] = startdate
    parameters['enddate'] = enddate
    parameters['output'] = 'csv'

    r = requests.get(url , parameters)

    print(r.text)
```

```python
getstock('TSLA','2016-01-01','2016-12-31')
```
